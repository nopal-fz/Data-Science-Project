```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = ('/content/Advertising Budget and Sales.csv')

df = pd.read_csv(dataset)
print(f'Full shape dataset is {df.shape}')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'TV Ad Budget ($)': 'TV', 'Radio Ad Budget ($)':'Radio', 'Newspaper Ad Budget ($)':'Newspaper','Sales ($)':'Sales'}, inplace=True)
df.head()
```

```python
figure = px.scatter(df, x='Sales', y='TV', size='TV',trendline='ols')
figure.show()
```

```python
figure = px.scatter(df, x='Sales', y='Radio', size='Radio',trendline='ols')
figure.show()
```

```python
figure = px.scatter(df, x='Sales', y='Newspaper', size='Newspaper',trendline='ols')
figure.show()
```

##### Find the correlation of each variable

```python
correlation = df.corr()
correlation['Sales'].sort_values(ascending=False)
```

##### Create a modeling and prediction

```python
x = np.array(df.drop(['Sales'], axis=1))
y = np.array(df['Sales'])

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

##### Do a perform test on the model

```python
model = LinearRegression()
model.fit(xtrain, ytrain)

print(model.score(xtest, ytest))
```

##### Do a prediction

```python
# Features = TV, Radion, Newspaper
predict_features = np.array([[230.1, 37.8, 69.2]])
print(model.predict(predict_features))
```
